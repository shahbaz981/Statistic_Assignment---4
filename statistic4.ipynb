```python
#Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

'''[Q1-answer:-]
The Probability Mass Function (PMF) and Probability Density Function (PDF) are both mathematical concepts
used in probability and statistics to describe the distribution of a random variable, but they are applicable
in different contexts.

    (1) Probability Mass Function (PMF):
            * The PMF is used for discrete random variables, which take on a finite or countably infinite 
                number of possible values.
                
            * It gives the probability that a discrete random variable is exactly equal to some value.
            
            * Mathematically, the PMF is defined as P(X = x), where X is the random variable and x is a
                specific value that X can take.
                
            * The sum of the probabilities for all possible values of the random variable must equal 1.
            
            * An example of a discrete random variable is rolling a fair six-sided die. The PMF for this 
                scenario would be a function that assigns a probability to each possible outcome: P(X = 1),
                P(X = 2), ..., P(X = 6), where X represents the outcome of the die roll.
                
    (2) Probability Density Function (PDF):
            * The PDF is used for continuous random variables, which can take on any value within a given
                range.
                
            * It represents the likelihood of a continuous random variable falling within a particular range
                of values.
                
            * Unlike the PMF, the PDF itself does not directly give probabilities but rather probability
                densities. The probability of a continuous random variable falling within a specific interval
                is given by the integral of the PDF over that interval.
                
            * The area under the curve of the PDF over any interval gives the probability of the variable
                falling within that interval.
                
            * For example, the PDF of a standard normal distribution (bell curve) is given by the
                formula f(x) = (1/√(2πσ^2)) * exp(-(x-μ)^2/(2σ^2)), where μ is the mean and σ^2 is the
                variance. This function describes the distribution of continuous values of a standard normal
                random variable.
'''
```

```python
#Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

'''[Q2-answer:-]
The Cumulative Density Function (CDF) is a fundamental concept in probability theory and statistics.
It provides the probability that a random variable X will take on a value less than or equal to a specific
value x. Mathematically, it is denoted as F(x) = P(X ≤ x).

Here's an explanation with an example:

Example:-
    Let's consider the example of rolling a fair six-sided die. The CDF for this scenario would describe 
    the probability of obtaining a value less than or equal to a specific outcome.

    For a fair six-sided die, each outcome has an equal probability of 1/6. Therefore, the CDF for rolling
    a fair six-sided die can be represented as:

                F(x) = 0 for x < 1
                F(x) = 1/6 for 1 ≤ x < 2
                F(x) = 2/6 for 2 ≤ x < 3
                F(x) = 3/6 for 3 ≤ x < 4
                F(x) = 4/6 for 4 ≤ x < 5
                F(x) = 5/6 for 5 ≤ x < 6
                F(x) = 1 for x ≥ 6

    This CDF indicates that if we roll the die, the probability of getting a value less than or equal to 1 is 1/6,
    less than or equal to 2 is 2/6, and so on, until reaching 1 for values greater than or equal to 6.
    
---------------
Why CDF is Used:-
        (1) Easy Interpretation: The CDF provides a straightforward interpretation of the cumulative
            probability distribution. It tells us the probability of a random variable being less than or 
            equal to a certain value.

        (2) Calculation of Percentiles: Percentiles, quartiles, and other statistical measures can be derived
            directly from the CDF. For instance, the median (50th percentile) is the value of x for which 
            F(x) = 0.5.

        (3) Comparison of Distributions: CDFs offer a convenient way to compare different probability
            distributions. By comparing their shapes and behaviors, one can infer similarities and differences
            between various random variables.

        (4) Integration with Continuous Distributions: For continuous distributions, the CDF is essential for
            calculating probabilities over intervals. This integration is crucial in many statistical
            analyses and applications.
'''
```

```python
#Q3: What are some examples of situations where the normal distribution might be used as a model?
#    Explain how the parameters of the normal distribution relate to the shape of the distribution.

'''[Q3-answer:-]
The normal distribution, also known as the Gaussian distribution, is a fundamental concept in statistics and
probability theory. It is widely used to model a variety of natural phenomena and real-world data due to
its mathematical properties and applicability to many situations. Here are some examples of situations
where the normal distribution might be used as a model:

    (1) Biological Traits: Many biological traits, such as height, weight, blood pressure, and IQ scores,
        often follow a normal distribution within a population.

    (2) Measurement Errors: In fields like engineering and physics, measurement errors tend to follow a
        normal distribution, assuming the errors are small and random.

    (3) Financial Data: Returns on investments, stock prices, and financial market fluctuations often
        exhibit normal distribution characteristics, particularly over short time intervals.

    (4) Psychological Variables: Psychological variables like aptitude test scores, personality traits,
        and reaction times in experiments can often be modeled using the normal distribution.

    (5) Environmental Factors: Natural phenomena like rainfall, wind speed, and temperature can sometimes 
        be approximated by a normal distribution under certain conditions.

-----------------------------------------------------------------------------------------------------------
Now, let's delve into how the parameters of the normal distribution relate to the shape of the distribution.

    The normal distribution is characterized by two parameters: the mean (μ) and the standard deviation (σ).
    These parameters define the center and spread of the distribution, respectively.

        (1) Mean (μ): The mean is the central value around which the data is distributed. It determines the
            location of the peak of the distribution. When the mean changes, the entire distribution shifts
            left or right along the x-axis. For example, increasing the mean shifts the distribution to the
            right, while decreasing it shifts the distribution to the left.

        (2) Standard Deviation (σ): The standard deviation measures the spread or dispersion of the data
            points around the mean. A larger standard deviation indicates that the data points are more
            spread out from the mean, resulting in a wider and flatter distribution. Conversely, a smaller 
            standard deviation leads to a narrower and taller distribution. Changes in the standard
            deviation directly affect the width of the bell curve: increasing the standard deviation
            spreads out the curve, while decreasing it makes the curve narrower.
'''
```

```python
#Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?

'''[Q1-answer:-]
The Bernoulli distribution is a probability distribution that models a random experiment with two possible
outcomes, typically labeled as success and failure, each with a certain probability. It is named after
Swiss mathematician Jacob Bernoulli, who introduced it in the late 17th century.

In a Bernoulli distribution:
    * There are only two possible outcomes, conventionally denoted as 1 (success) and 0 (failure).
    * There is a fixed probability p of success and 1−p of failure in each trial.
    * It is a discrete distribution.
    
A simple example of a Bernoulli experiment is flipping a fair coin, where getting heads might be considered
a success (1) and getting tails a failure (0). Suppose the probability of getting heads is p = 0.5, then
the Bernoulli distribution can model the probability of getting heads or tails in a single flip of the coin.

-------------------------------------------------------------------------
The Bernoulli distribution and the binomial distribution are closely related, but they describe different
scenarios and have distinct characteristics:

(1) Bernoulli Distribution:
        * Describes a single trial or experiment with two possible outcomes, often labeled as success (1)
          and failure (0).
        * Has a single parameter p, which represents the probability of success in a single trial.
        * The random variable X follows a Bernoulli distribution if it takes the value 1 with probability 
          p and the value 0 with probability 1−p.
        * Example: Flipping a biased coin once, where the outcome of heads could be considered a success
          and tails a failure.
          
(2) Binomial Distribution:
        * Describes the number of successes in a fixed number of independent Bernoulli trials.
        * Has two parameters: n, the number of trials, and p, the probability of success in each trial.
        * The random variable X follows a binomial distribution if it represents the number of successes in 
           n independent Bernoulli trials, each with probability p.
        * The probability mass function (PMF) of the binomial distribution gives the probability of
          obtaining a specific number of successes in n trials.
        * Example: Counting the number of heads obtained when flipping a biased coin n times.
'''
```

```python
#Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is
#    normally distributed, what is the probability that a randomly selected observation will be greater than 60?
#    Use the appropriate formula and show your calculations.

'''[Q6-answer:-]
To find the probability that a randomly selected observation from a normally distributed dataset with a mean
of 50 and a standard deviation of 10 will be greater than 60, we need to use the standard normal distribution
and the z-score formula.

First, we calculate the z-score for x = 60 using the formula:

                            z = (x−μ)/σ

        Where:
                x is the value we're interested in (60 in this case),
                μ is the mean of the dataset (50),
                σ is the standard deviation of the dataset (10).
                
                
            z = (60−50)/10 = 10/10 = 1


    Now, we look up the probability corresponding to a z-score of 1 using a standard normal distribution table or
    a calculator. The probability that a randomly selected observation will be greater than 60 is the area under
    the standard normal curve to the right of the z-score of 1.

    From a standard normal distribution table or calculator, we find that the probability corresponding to a
    z-score of 1 (the area to the right of 1) is approximately 0.1587.

    So, the probability that a randomly selected observation will be greater than 60 is approximately 0.1587,
    or 15.87%.

    This calculation indicates that there is approximately a 15.87% chance that a randomly selected observation
    from the dataset will be greater than 60.
'''
from scipy.stats import norm

# Mean and standard deviation of the dataset
mean = 50
std_dev = 10

# Value of interest
x = 60

# Calculate z-score
z = (x - mean) / std_dev

# Calculate probability using the cumulative distribution function (CDF) of the standard normal distribution
# Since we're interested in values greater than 60, we use the complementary CDF (1 - CDF)
probability_greater_than_60 = 1 - norm.cdf(z)

print("Probability that a randomly selected observation will be greater than 60:", probability_greater_than_60)
```

```python
#Q8: What is the z score? State the importance of the z score.

'''[Q8-answer:-]
The z-score, also known as the standard score, is a statistical measure that quantifies the number of 
standard deviations a data point is from the mean of a dataset. It is calculated using the

        formula:-
                            z = (x−μ)/σ

            Where:
                    x is the individual data point,
                    μ is the mean of the dataset,
                    σ is the standard deviation of the dataset.
                
        The z-score indicates how many standard deviations an observation is above or below the mean.
        A positive z-score indicates that the observation is above the mean, while a negative z-score
        indicates that the observation is below the mean. A z-score of 0 means the observation is exactly
        at the mean.
        
--------------------------------------------------------
The importance of the z-score lies in several key aspects:

        (1) Standardization: The z-score standardizes different datasets, allowing for meaningful comparison
            and interpretation across different scales and distributions. By expressing data points in terms
            of standard deviations from the mean, z-scores provide a common scale that facilitates comparison.

        (2) Identifying Outliers: Z-scores can be used to identify outliers in a dataset. Data points with
            z-scores that are significantly larger or smaller than 0 (typically beyond ±3 standard deviations)
            may be considered outliers and warrant further investigation.

        (3) Probability Calculation: Z-scores are commonly used in probability calculations, particularly in
            the context of the standard normal distribution. Z-scores represent the probability of obtaining
            a value less than or equal to a given value in a standard normal distribution, which is crucial
            for hypothesis testing and confidence interval estimation.

        (4) Data Transformation: Z-scores are often used in data transformation techniques, such as
            normalization and standardization, to prepare data for certain statistical analyses or modeling
            techniques. Transforming data into z-scores can help mitigate the influence of differences in
            scale and variance among variables.

        (5) Statistical Inference: In hypothesis testing, z-scores play a key role in determining whether a
            sample mean or proportion is significantly different from a hypothesized population mean or
            proportion. They are used to calculate test statistics and p-values, which inform decisions about
            accepting or rejecting null hypotheses.
'''
```

```python
#Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

'''[Q9-answer:-]
The Central Limit Theorem (CLT) is a fundamental concept in probability theory and statistics. It states that
the sampling distribution of the sample mean (or the sum) of a sufficiently large number of independent and
identically distributed random variables will approximate a normal distribution, regardless of the shape of
the original population distribution. In simpler terms, it implies that when we repeatedly sample from a
population and calculate the mean of each sample, those means will tend to follow a normal distribution,
regardless of the shape of the population distribution.

    --> The significance of the Central Limit Theorem lies in its broad applicability and its implications
        for statistical inference. Some key points about its significance include:

            (1) Robustness: The CLT provides a powerful tool for making inferences about population
                parameters, even when the underlying population distribution is unknown or not normally
                distributed. This is particularly valuable in real-world applications, where population
                distributions may be complex or not well understood.

            (2) Inference: It forms the basis for many statistical methods and hypothesis tests, such as
                confidence intervals and hypothesis testing for population means. These methods rely on the
                normality of the sampling distribution of the sample mean, which is assured by the CLT under
                appropriate conditions.

            (3) Sampling Theory: The CLT is foundational in sampling theory, allowing statisticians to make
                probabilistic statements about sample statistics and their relationship to population 
                parameters.

            (4) Quality Control: In fields such as manufacturing and quality control, where repeated sampling
                and measurement are common, the CLT provides a theoretical basis for understanding the
                behavior of sample statistics and establishing control limits.

            (5) Econometrics and Finance: In econometrics and finance, the CLT is used extensively in
                modeling and forecasting, particularly in the context of time series analysis and portfolio
                management.

            (6) Statistical Estimation: The CLT underpins methods of statistical estimation, such as maximum
                likelihood estimation and least squares estimation, by justifying the use of normality
                assumptions in large sample inference.
'''
```
